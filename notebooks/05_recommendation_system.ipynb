# Step 5: Product Recommendation System

This notebook builds a simple item-based recommendation system using
purchase history and cosine similarity.


### IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
```

### LOAD DATASET

```python
df = pd.read_csv('../data/cleaned_online_retail.csv')
df.head()
```

### CREATE CUSTOMER-PRODUCT MATRIX

```python
pivot_table = df.pivot_table(
    index='CustomerID',
    columns='Description',
    values='Quantity',
    aggfunc='sum',
    fill_value=0
)

pivot_table.head()
```

### COMPUTE COSINE SIMILARITY

```python
item_similarity = cosine_similarity(pivot_table.T)
```

```python
item_similarity_df = pd.DataFrame(
    item_similarity,
    index=pivot_table.columns,
    columns=pivot_table.columns
)

item_similarity_df.head()
```

### RECOMMENDATION FUNCTION

```python
def recommend_products(product_name, top_n=5):
    if product_name not in item_similarity_df.columns:
        return "Product not found."
    
    similar_scores = item_similarity_df[product_name].sort_values(ascending=False)
    recommended_products = similar_scores.iloc[1:top_n+1].index.tolist()
    
    return recommended_products
```

### TEST THE RECOMMENDER

```python
recommend_products("WHITE HANGING HEART T-LIGHT HOLDER")
```

### Recommendation System Insights

- An item-based collaborative filtering approach was implemented.
- Cosine similarity was used to measure similarity between products.
- The system recommends products frequently purchased together.
- This approach is simple, efficient, and suitable for large-scale e-commerce data.
